```python
# 플레이리스트 제목 중복 merge
# train # 115071
train['plylst_title'].duplicated().sum()
```

```python
# train[train['plylst_title'].duplicated()]
train[train['plylst_title'] == '미리 메리 크리스마스']
```

```python
def f(x):
  d = {}
  d['tags'] = list(set(sum(x.tags, [])))
  d['id'] = x['id'].max()
  d['plylst_title'] = x['plylst_title'].max()
  d['songs'] = list(set(sum(x.songs, [])))
  d['like_cnt'] = x['like_cnt'].sum()
  d['updt_date'] = x['updt_date'].max()
  return pd.Series(d, index=['tags', 'id', 'plylst_title', 'songs', 'like_cnt', 'updt_date'])
unique_train = train.groupby(by='plylst_title').apply(f)
unique_train[unique_train['plylst_title'] == '부활']
```

```python
unique_train[unique_train['plylst_title'] == '미리 메리 크리스마스']
```

```python
# 플레이리스트 수록곡 한개 이하인 플레이리스트 drop
unique_train.drop(unique_train[unique_train['songs'].apply(len) <= 1].index, inplace=True)
```
